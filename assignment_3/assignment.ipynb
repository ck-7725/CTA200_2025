# Question 1
For each point in the complex plane $c = x + iy$, with $-2 < x < 2$ and $-2 < y < 2$, set $z_0 = 0$ and iterate the equation $z_{i + 1} = z_i^2 + c$. 
Note what happens to the $z_i$'s: some points will remain bounded in absolute value $|z|^2 = \Re(z)^2 + \Im(z)^2$, while others will run off to infinity. 
Make an image  in which your points $c$ that diverge are given one color and those that stay bounded are given another.
Make a second image where the points are coloured by a colourscale that indicates the iteration number at which the given point diverged.

For this question, put the code that does the iteration in a function and place this function in a separate .py file which you import in your .ipynb.
Perform the plots in the notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
real = np.linspace(-2, 2, num=1000)
imag = np.linspace(-2, 2, num=1000)

plane = real[np.newaxis, :] + imag[:, np.newaxis] * 1j
```

```python
def iterate_binary(c, z0=0, iterations=50):
    z = z0
    for _ in range(iterations):
        z = z ** 2 + c
        if np.isinf(abs(z)):
            return False
    return True
```

```python
bounded = np.empty((len(plane), len(plane[0])))

for row in range(len(plane)):
    for col in range(len(plane[row])):
        bounded[row][col] = iterate_binary(plane[row][col])
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(bounded, cmap="binary")
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
```

```python
def iterate_scale(c, z0=0, iterations=50):
    z = z0
    for i in range(iterations):
        z = z ** 2 + c
        if np.isinf(abs(z)):
            return i
    return -1
```

```python
iter_num = np.empty((len(plane), len(plane[0])))

for row in range(len(plane)):
    for col in range(len(plane[row])):
        iter_num[row][col] = iterate_scale(plane[row][col])
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(iter_num, cmap='inferno')
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
```

```python
# 1. code up the equations, using a function definition, with a proper docstring (inside triple quotes)

def lorenz(t, W, sigma=10.0, r=28.0, b=8.0/3.0):
    """This function represents Lorenz's Equations 25, 26, and 27.
    
    Parameters
    ----------
    W : list
        Initial amplitudes, W = [X, Y, Z]
    sigma : float
        Prandtl number
    r : float
        Rayleigh number
    b : float
        Dimensionless length scale
    
    Returns
    -------
    W_dot : list
        Next iteration of W based on Lorenz's equations and the given parameters
    """
    X, Y, Z = W
    
    X_dot = -sigma * (X - Y)
    Y_dot = r * X - Y - X * Z
    Z_dot = -b * Z + X * Y
    W_dot = [X_dot, Y_dot, Z_dot]
    return W_dot
```

```python
from scipy.integrate import solve_ivp

W0 = [0.0, 1.0, 0.0]
t_span = (0, 60)
t_eval = np.linspace(*t_span, 6000)

# Solve Lorenz system
sol = solve_ivp(lorenz, t_span, W0, t_eval=t_eval)
```

```python
# Reproducing Figure 1
fig, ax = plt.subplots(3, 1, figsize=(6, 6))

ax[0].plot(sol.t[0:1000], sol.y[1][0:1000])
ax[1].plot(sol.t[1001:2000], sol.y[1][1001:2000])
ax[2].plot(sol.t[2001:3000], sol.y[1][2001:3000])

fig.tight_layout()
```

```python
# Reproducing Figure 2

fig, ax = plt.subplots(2, 1, figsize=(6, 12))

ax[0].plot(sol.y[1], sol.y[2])
ax[1].plot(sol.y[1], sol.y[0])

fig.tight_layout()
```

```python
# Solving with slightly different initial conditions

W_prime = [W0[0] + 0., W0[1] + 1e-8, W0[2] + 0.]

sol_prime = solve_ivp(lorenz, t_span, W_prime, t_eval=t_eval)
```

```python
# Plotting difference

plt.figure(figsize=(8, 6))
dist = abs(sol_prime.y - sol.y)
plt.semilogy(sol.t, dist[0])
```
